```python
import re
import os
import csv

def extract_county_to_city_or_district(file_path):
    results = []

    # Extract year from filename like "diff/1980-1981.diff"
    year_match = re.search(r'(\d{4}-\d{4})', file_path)
    year = year_match.group(1) if year_match else "Unknown"

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()

            # Case 1: + or = lines using <
            match1 = re.match(r'^([+=])(\d+)\s+([^\s<>=]+[Â∏ÇÂå∫])<([^<>=]*Âéø[^<>=]*)', line)
            # Case 2: - lines using >
            match2 = re.match(r'^(-)(\d+)\s+([^>Âéø]*Âéø)[>]([^\s<>=]+[Â∏ÇÂå∫])', line)

            if match1:
                symbol, code, target, source = match1.groups()
                old_code = "NA" if symbol == "+" else code
                new_code = code if symbol == "+" else "NA"
                results.append([
                    year,
                    source.strip(),
                    old_code,
                    target.strip(),
                    new_code,
                    symbol
                ])

            elif match2:
                symbol, code, source, target = match2.groups()
                results.append([
                    year,
                    source.strip(),
                    code,
                    target.strip(),
                    "NA",
                    symbol
                ])

    return results

# üîÅ Process all .diff files in the "diff" folder
all_results = []
diff_folder = "diff"

for filename in os.listdir(diff_folder):
    if filename.endswith(".diff"):
        file_path = os.path.join(diff_folder, filename)
        print(f"Processing: {file_path}")
        changes = extract_county_to_city_or_district(file_path)
        all_results.extend(changes)

# üíæ Save to CSV
output_file = "change in administration.csv"
with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["year", "old name", "old code", "new name", "new code", "change type"])
    writer.writerows(all_results)

print(f"\n‚úÖ Saved all results to: {output_file}")
```
